```python
import csv
from tkinter import *

# the following two are field names for the dictionary and the dictionary file path where the dictionaries 
#will be stored.
fieldnames = ['NEW TERM', 'MEANING', 'SYNTAX', 'EXAMPLE']
dictionary_path = "interactive_dictionary.csv"

# this message will be displayed every time a dictionary is updated which will be after clicking 
# the Updated Dictionary button.
update_message = "Dictionary updated!"

# see what the function does. 
def clear_all(): 
    '''this function clears all the values in the widgets listed when invoked.'''
    text2.delete(1.0,END)
    text3.delete(1.0,END)
    text4.delete(1.0,END)
    text5.delete(1.0,END)
    
def clear_update(): 
    '''this function clears or deletes the Update message when invoked.'''
    message1.delete(1.0,END)
    
def all_values():
    '''this function fetches the values from the text boxes and then appends them in the form of a 
    dictionary to a csv file.'''
    
    # here we fetch the values entered into the text boxes.
    new_term = text2.get(1.0, END).strip()  
    meaning = text3.get(1.0, END).strip()
    syntax = text4.get(1.0, END).strip()
    example = text5.get(1.0, END).strip()
    print (type(new_term))
    
    # the fetched values are entered into a list. 
    values = []
    values.append(new_term)
    values.append(meaning)
    values.append(syntax)
    values.append(example)
    
    # a dictionary row is constructed from them. 
    dic_row = dict(zip(fieldnames, values))
    
    # we are opening the file in append mode so we can keep adding to it. 
    with open(dictionary_path, "a") as csvfile: 
        
        # let us create our dict writer
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames) ## writing column names
        
        # let us see where the current position of the cursor is. 
        current_position = csvfile.tell()
        
        # if the current position of the cursor is at 0, then write a header, otherwise, just write a row.
        if current_position == 0: 
            writer.writeheader()
            writer.writerow(dic_row)
        else: 
            writer.writerow(dic_row)
    
    # update message is displayed.     
    message1.insert(END, update_message)
    clear_all() #clear all function is invoked clearing all the boxes. 

def read_values():
    '''this function returns a searched term and the entire dictionary row associated with that term.'''
    text1.delete(1.0, END) # the format for text box is text.delete(1.0, END) for other widgets maybe different.
    search_value = text6.get(1.0, END).strip() # we are using strip() method otherwise it will add the \n
    with open(dictionary_path) as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            if row['NEW TERM'] == search_value: 
                text1.insert(END, "TERM: {}\nMEANING: {}\nSYNTAX: {}\nEXAMPLE: {}".format(row['NEW TERM'], row['MEANING'], row['SYNTAX'], row['EXAMPLE']))




root = Tk()
root.title("Interactive Dictionary")
root.configure(background = '#999966')

lab_1 = Label(root, text=fieldnames[0], height=2, width = 15, background="#996699")
lab_1.grid(row=0, column=0, padx=10, pady=10)
#text2_value = get(start, end=None)
text2 = Text(root, height=2, width=40, background="#e9e9e2")
text2.grid(row=0, column=1)

lab2 = Label(root, text=fieldnames[1], height=2, width = 15, background="#996699")
lab2.grid(row=1, column=0)

text3 = Text(root, height=2, width=50, background="#e9e9e2")
text3.grid(row=1, column=1)

lab3 = Label(root, text=fieldnames[2], height=2, width = 15, background="#996699")
lab3.grid(row=2, column=0, pady=10)

text4 = Text(root, height=2, width=50, background="#e9e9e2")
text4.grid(row=2, column=1)

lab4 = Label(root, text=fieldnames[3], height=2, width = 15, background="#996699")
lab4.grid(row=3, column=0)

text5 = Text(root, height=2, width=80, background="#e9e9e2")
text5.grid(row=3, column=1)

btn_clr = Button(root, text="CLEAR UPDATE MESSAGE", height = 1, width=18, command=clear_update, bg="blue")
btn_clr.grid(row=5, column=0, pady=10, padx=20)

btn1 = Button(root, text="Update Dictionary!", command=all_values, bg="blue")
btn1.grid(row=5, column=1)
message1 = Text(root, height=1, width = 22, background="#e9e9e2")
message1.grid(row=5, column=2, padx = 30)

btn2 = Button(root, text="SEARCH TERM", height=1, width=12, command=read_values, bg="blue")
btn2.grid(row=8, column=0, pady=20)

text6 = Text(root, height=2, width=40, background="#e9e9e2")
text6.grid(row=8, column=1)

lab5 = Label(root, text="RESULT BOX", relief = RAISED, background="#996699")
lab5.grid(row=9, column=0, pady=20)
text1 = Text(root, height=20, background="#e9e9e2")
text1.grid(row=9, column=1, pady=10, padx=10)

root.mainloop()
```
